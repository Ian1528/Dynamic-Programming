- schedule the class with the most demand first
- scheudle all the students and classes first, then assign the teachers.

Two ways to think about cost/optimal next choice:
- the class with the most studnets signed up for it (do this until every block fo the school day is full)
- the class with the least amount of students that would have a conflict with an existing, scheduled class

### Assigning students to courses

Start with the course with the least # of sections.

Then, iterate over the sections, and for each student in that course:
- assign the students based on their scheudles, which has two parts.
    - Assign students with the least other classes also taught in this block (minimize conflict of schedules). If there is overlap, assign the students with the most # of available sections combined between their other courses (in theory, they are the most flexible)

- conflict resolution: if there are no available courses for a student:
    - first, try and move one of their other courses to accomodate.
    - if not possible, break the max class size limit and add a penalty
    - if still infeasible, make a note of that (sometimes conflicts are unavoidable depending on teacher schedules)

```python

def recursively_assign_students_using_max_flow(courses: list[Course], give_output = False) -> list[Course]:
    """
    Assigns students to sections using a max flow algorithm.

    Args:
        courses: a list of the courses to assign
    Returns:
        A list of courses that still need to be assigned students due to shuffeling.
    """
    required_courses = [course for course in courses if not course.is_elective]
    elective_courses = [course for course in courses if course.is_elective]

    sorted_required_courses = sorted(required_courses, key=lambda x: len(x.sections))
    sorted_elective_courses = sorted(elective_courses, key=lambda x: len(x.sections))

    sorted_courses = sorted_required_courses + sorted_elective_courses

    courses_remaining: list[Course] = []
    for course in sorted_courses:
        ideal_students_per_section = math.ceil(len(course.students)/len(course.sections))

        # construct networkx graph, where each student is connected to each section they can take
        # each section is connected to a sink node
        # each student is connected to a source node
        G = nx.DiGraph()
        G.add_node("source")
        G.add_node("sink")

        blocks_to_sections: dict[str, Section] = {}
        for section in course.sections:
            G.add_node(section)
            G.add_edge(section, "sink", capacity=ideal_students_per_section)
            blocks_to_sections[section.block] = section

        unassigned_students: list[Student] = []
        students_to_assign: list[Student] = []

        students_ranked = sorted(course.students, key=cmp_to_key(compare_students))

        # somehow, the order of the students changes the way they are assigned
        for student in students_ranked:
            # if the student is already assigned to a course in this section, skip them
            current_section = student.get_course_section(course)
            if current_section != None:
                # instead of changing capacity, what if we added the student to the graph with zero weight and infinite capacity?
                print(student, "already assigned to section", current_section.block, "for course", course.name)
                if G.edges()[(current_section, "sink")]["capacity"] == 0:
                    print("Zero capacity! Students already assigned to this section", len(current_section.students), "Ideal students per section", ideal_students_per_section)
                else:   
                    G.edges()[(current_section, "sink")]["capacity"] -= 1

            else:
                students_to_assign.append(student)
                sections_for_student = student.get_available_sections_with_saturation(course)
                if len(sections_for_student) == 0:
                    unassigned_students.append(student)
                    continue
                G.add_node(student.name)
                G.add_edge("source", student.name, capacity=1)
                for section, saturation in sections_for_student.items():
                    # infinite capacity, weight is the saturation
                    G.add_edge(student.name, section, weight=saturation)
            
        # find the max flow min cost in the graph
        flow_dict = nx.max_flow_min_cost(G, "source", "sink")
        # assign the students to the sections based on the flow dict
        for student in students_to_assign:
            if student in unassigned_students:
                continue
            best_section = max(flow_dict[student.name], key=lambda x: flow_dict[student.name][x])

            if student.get_course_section(course) != None and student.get_course_section(course) != best_section:
                print("Flow network:")
                for key, value in flow_dict.items():
                    print(key, "->", value)
                print("Edges are")
                for edge in G.edges(data=True):
                    print(edge)

                print("Student", student.name, "already assigned to section", student.get_course_section(course).block, "for course", course.name)
                print("Best section is", best_section.block, "for course", course.name) 

                raise Exception("Student already assigned to a section, but the best section is different.")
            if flow_dict[student.name][best_section] == 0:
                unassigned_students.append(student)
                continue
            
            # student already assigned to this section, so skip them
            best_section.add_student(student)

        # not all students were able to be assigned to a section, so we need to try to assign them to the sections with the least number of students and shuffle things around
        if len(unassigned_students) > 0:
            for student in unassigned_students:
                sections_sorted_by_size = sorted(course.sections, key=lambda x: len(x.students))
                
                assigned = False
                
                # try to assign the student to the section with the least number of students
                for min_size_section in sections_sorted_by_size:
                    if student.schedule[min_size_section.block] == None:
                        min_size_section.add_student(student)
                        assigned = True
                        break
                    if student.reassign_block(min_size_section.block, [min_size_section.block], 0):
                        min_size_section.add_student(student)
                        assigned = True
                        break
                # if the student is not able to be assigned to any of the sections, they cannot take this course
                if not assigned:                    
                    # get student course preferences
                    for course_requests in student.course_requests:
                        has_alternative = False
                        try:
                            course_preference_rank = course_requests.index(course)
                            if course_preference_rank + 1 < len(course_requests):
                                has_alternative = True
                                next_course = course_requests[course_preference_rank + 1]
                            break
                        # haven't found the course in the list yet, so continue searching
                        except ValueError:
                            continue
                    
                    if not has_alternative:
                        continue
                        # print(f"Student {student.name} does not have any ranked alternative electives available.") 
                    else:
                        print(f"Changing {student.name} top choice from {course.name} to {next_course.name}")
                        course.remove_student(student)
                        next_course.add_student(student)
                        courses_remaining.append(course)
        if give_output and course.name == "MAT50":
            print("Flow network:")
            for key, value in flow_dict.items():
                print(key, "->", value)
            print("Edges are")
            for edge in G.edges(data=True):
                print(edge)
            print("Students in each section:")
            for section in course.sections:
                print(section.block, "has", len(section.students), "students")
                for student in section.students:
                    print(student.name, "is in block", student.schedule[section.block])
                print("**")
    while len(courses_remaining) > 0:
        courses_remaining = recursively_assign_students_using_max_flow(courses_remaining)
    
    return courses_remaining
def compare_students(s1: Student, s2: Student):
    s1_block_overlap = 0
    for course in s1.get_all_first_choice_courses():
        for section in course.sections:
            if section.block == section.block:
                s1_block_overlap += 1

    s2_block_overlap = 0
    for course in s2.get_all_first_choice_courses():
        for section in course.sections:
            if section.block == section.block:
                s2_block_overlap += 1
    
    if s1_block_overlap != s2_block_overlap:
        return s1_block_overlap - s2_block_overlap # if s1 has more overlap, it is less desirable to place in this block

    else:
        return s1.get_flexibility() - s2.get_flexibility() # students with less flexibility are more desirable to place in this block

def rollout(required_courses: list[Course], elective_courses: list[Course]):
    required_courses_sorted = sorted(required_courses, key=lambda x: len(x.sections))
    elective_courses_sorted = sorted(elective_courses, key=lambda x: len(x.sections))

    sorted_courses = required_courses_sorted + elective_courses_sorted
    
    all_courses_copy = deepcopy(sorted_courses)
    overall_best_cost = math.inf

    for i in range(len(all_courses_copy)):
        course = all_courses_copy[i]

        students_ranked = sorted(course.students, key=cmp_to_key(compare_students))
        for j in range(len(students_ranked)):
            # print("Student", j, "of", len(students_ranked), "for course", course.name, "iteration", i)
            student = students_ranked[j]

            best_state_section = None
            best_state_cost = math.inf

            current_test_state = deepcopy(all_courses_copy)
            best_test_state = None

            for k in range(len(course.sections)):
                # print("Testing section", k, "of", len(course.sections), "for student iteration", j, "for", course.name, "iteration", i)
                # rollout with copies of the original state
                section = current_test_state[i].sections[k]
                current_students_ranked = sorted(current_test_state[i].students, key=cmp_to_key(compare_students))
                current_student = current_students_ranked[j]
                
                
                if not section.is_full() and current_student.reassign_block(section.block, [section.block], 0):
                    section.add_student(current_student)

                    # if k == 1 and student.name == "Student 42 Grade 9" and course.name == "MAT10":
                    #     print("OUTPUT FOR CASE WHERE COST IS 45 but the section is the same ast the best cost")
                    #     recursively_assign_students_using_max_flow(current_test_state[i:], give_output=True)
                    #     for course in current_test_state[:3]:
                    #         print(course.name, "has", len(course.students), "students")
                    #         for section in course.sections:
                    #             print("Section", section.block, "has", len(section.students), "students")
                    #             for student in section.students:
                    #                 print(student.name, "is in block", student.schedule[section.block])
                    #     MATH_10_COPY = deepcopy(current_test_state)
                    # else:
                    recursively_assign_students_using_max_flow(current_test_state)
                    cost = cost_of_final_schedules_from_courses(current_test_state)
                    if course.name == "MAT50" and student.name == "Student 46 Grade 12":
                        print("for", section, "student", student.name, "cost is", cost)
                        print("in the optimal solution it goes in section 3")
                    # update the best state if the cost is lower
                    if cost <= best_state_cost:
                        best_state_cost = cost
                        best_state_section = k
                        # best_test_state = deepcopy(current_test_state)
                    # reset the courses to their original state
                    current_test_state = deepcopy(all_courses_copy)
                else:
                    print("Unable to assign. Is section full?", section.is_full(), "Is student already assigned?", current_student.get_course_section(course) != None, current_student.schedule[section.block])
            # assign the student based on the best result. TODO: check the case where the outcome is None
            if best_state_section == None:
                print("Unable to assign student", j, "to any sections for", course.name)
                continue
            else:
                # what about the case where the student schedule was shuffled using reassign_block?
                best_section = course.sections[best_state_section]
                assigned = student.reassign_block(best_section.block, [best_section.block], 0)

                best_section.add_student(student)
                for i2 in range(len(student.schedule.values())):
                    for j2 in range(i2+1, len(student.schedule.values())):
                        if student.schedule[i2] != None and student.schedule[j2] != None and student.schedule[i2].name == student.schedule[j2].name:
                            print("Student", student.name, "is in the same course twice!", student.schedule)
                # print("Best cost so far after adding student", student, "is", best_state_cost)

                best_current_test_state = deepcopy(all_courses_copy)
            if best_state_cost <= overall_best_cost:
                overall_best_cost = best_state_cost
                overall_best_state = deepcopy(all_courses_copy)
                print("Overall best cost so far is being updated to", overall_best_cost)
                # print("Regaining best cost from the best state")
                # recursively_assign_students_using_max_flow(overall_best_state, give_output=False)
                # cost_after = cost_of_final_schedules_from_courses(overall_best_state)
                # if cost_after != overall_best_cost:
                #     print("Cost after is not the same as overall best cost. This is a problem.")
                #     print("comparing the two states")
                #     print(compare_two_master_schedules(overall_best_state, best_test_state))
                #     raise ValueError("Cost after is not the same as overall best cost. This is a problem.")
                # print("Best cost should be the same as before", cost_of_final_schedules_from_courses(overall_best_state))
                # print("Cost of besttesttate", cost_of_final_schedules_from_courses(best_test_state))
                # print("Comparing test_overall_best and best_test_state")
                # print(compare_two_master_schedules(overall_best_state, best_test_state))
            elif best_state_cost > overall_best_cost:
                print("Best cost is worse than overall best cost. Stopping search.")
                print("Iteration number", i, "for course", course.name, "section", best_state_section, "cost", best_state_cost)

                # compare the best state with the overall best state
                print("Best state original is")
                for course in overall_best_state:
                    print(course.name, "has", len(course.students), "students")
                    for section in course.sections:
                        print("Section", section.block, "has", len(section.students), "students")
                        for student in section.students:
                            print(student.name, "is in block", student.schedule[section.block])
                
                print("Best state after this iteration is")
                for course in best_current_test_state:
                    print(course.name, "has", len(course.students), "students")
                    for section in course.sections:
                        print("Section", section.block, "has", len(section.students), "students")
                        for student in section.students:
                            print(student.name, "is in block", student.schedule[section.block])
                            
                print("Getting the best cost again from the best state")
                recursively_assign_students_using_max_flow(overall_best_state, give_output=True)
                best_cost = cost_of_final_schedules_from_courses(overall_best_state)
                print("Best cost is ", best_cost)
                # for course in overall_best_state[:3]:
                #     print(course.name, "has", len(course.students), "students")
                #     for section in course.sections:
                #         print("Section", section.block, "has", len(section.students), "students")
                #         for student in section.students:
                #             print(student.name, "is in block", student.schedule[section.block])

                print("******************8")
                print("Getting the best cost again from the current state")
                recursively_assign_students_using_max_flow(best_current_test_state, give_output=True)
                # for course in test_copy[:3]:
                #     print(course.name, "has", len(course.students), "students")
                #     for section in course.sections:
                #         print("Section", section.block, "has", len(section.students), "students")
                #         for student in section.students:
                #             print(student.name, "is in block", student.schedule[section.block])
                # print("COMPARING MATH10 best and OVERALL BEST")
                # print(compare_two_master_schedules(overall_best_state, MATH_10_COPY))

                print("COMPARING AFTER ADDING EVERYTHING")
                print(compare_two_master_schedules(overall_best_state, best_current_test_state))
                return
            
reset_student_assignments()
rollout(all_courses_required, all_courses_electives)
```

```python
def recursively_assign_students_using_max_flow(courses: list[Course], current_student: Student = None, section_to_assign: Section = None, give_output = False) -> list[Course]:
    """
    Assigns students to sections using a max flow algorithm.

    Args:
        courses: a list of the courses to assign
    Returns:
        A list of courses that still need to be assigned students due to shuffeling.
    """
    required_courses = [course for course in courses if not course.is_elective]
    elective_courses = [course for course in courses if course.is_elective]

    sorted_required_courses = sorted(required_courses, key=lambda x: len(x.sections))
    sorted_elective_courses = sorted(elective_courses, key=lambda x: len(x.sections))

    sorted_courses = sorted_required_courses + sorted_elective_courses

    courses_remaining: list[Course] = []
    for course in sorted_courses:
        ideal_students_per_section = math.ceil(len(course.students)/len(course.sections))

        # construct networkx graph, where each student is connected to each section they can take
        # each section is connected to a sink node
        # each student is connected to a source node
        G = nx.DiGraph()
        G.add_node("source")
        G.add_node("sink")

        blocks_to_sections: dict[str, Section] = {}
        for section in course.sections:
            G.add_node(section)
            G.add_edge(section, "sink", capacity=ideal_students_per_section)
            blocks_to_sections[section.block] = section

        unassigned_students: list[Student] = []
        students_to_assign: list[Student] = []

        students_ranked = sorted(course.students, key=cmp_to_key(compare_students))

        # somehow, the order of the students changes the way they are assigned
        count = 0
        for student in course.students:
            count += 1
            if course.name == "MAT50" and give_output:
                print("Student", student.name, "ranked", students_ranked.index(student), "of", len(students_ranked))
            # if the student is already assigned to a course in this section, skip them
            current_section = student.get_course_section(course)
            if current_section != None:
                G.add_node(student.name)
                G.add_edge("source", student.name, capacity=1)
                G.add_edge(student.name, current_section, weight=0)

                for section, saturation in student.get_available_sections_with_saturation(course).items():
                    # infinite capacity, weight is the saturation
                    # check if the edge exists first
                    if G.has_edge(student.name, section):
                        continue
                    else:
                        G.add_edge(student.name, section, weight=saturation)

                # instead of changing capacity, what if we added the student to the graph with zero weight and infinite capacity?
                # print(student, "already assigned to section", current_section.block, "for course", course.name)
                # if G.edges()[(current_section, "sink")]["capacity"] == 0:
                #     print("Zero capacity! Students already assigned to this section", len(current_section.students), "Ideal students per section", ideal_students_per_section)
                #     continue
                # G.edges()[(current_section, "sink")]["capacity"] -= 1

            # if we are on the student that we should be considering
            elif student == current_student and section_to_assign != None:
                # if the student is the one we are trying to assign, add them to the graph with zero weight and infinite capacity
                G.add_node(student.name)
                G.add_edge("source", student.name, capacity=1)
                G.add_edge(student.name, section_to_assign, weight=0)
                
                for section, saturation in student.get_available_sections_with_saturation(course).items():
                    # infinite capacity, weight is the saturation
                    # check if the edge exists first
                    if G.has_edge(student.name, section):
                        continue
                    else:
                        G.add_edge(student.name, section, weight=saturation)

            else:
                # students_to_assign.append(student)
                sections_for_student = student.get_available_sections_with_saturation(course)
                if len(sections_for_student) == 0:
                    unassigned_students.append(student)
                    continue
                G.add_node(student.name)
                G.add_edge("source", student.name, capacity=1)
                for section, saturation in sections_for_student.items():
                    # infinite capacity, weight is the saturation
                    G.add_edge(student.name, section, weight=saturation)
            
            students_to_assign.append(student)

        # find the max flow min cost in the graph
        flow_dict = nx.max_flow_min_cost(G, "source", "sink")
        # assign the students to the sections based on the flow dict
        if course.name == "MAT50" and give_output:
            print("FLOWDICT:")
            for k, v in flow_dict.items():
                print(k, ":", v)
        for student in students_to_assign:
            if course.name == "MAT50" and give_output and student.name == "Student 38 Grade 12":
                print("assigning special", student.name, "is unassigned?", student in unassigned_students)
                print("Their flow dict is ", flow_dict[student.name])
                print("their best section is ", max(flow_dict[student.name], key=lambda x: flow_dict[student.name][x]))

                # print out the entire graph
                for node in G.nodes():
                    print("Node", node, "has edges", G.edges(node))
                #print the flow dict
                for k, v in flow_dict.items():
                    print("Flow dict", k, ":", v)
            if student in unassigned_students:
                continue
            best_section = max(flow_dict[student.name], key=lambda x: flow_dict[student.name][x])
            if flow_dict[student.name][best_section] == 0:
                unassigned_students.append(student)
                continue
            
            # student already assigned to this section, so skip them
            best_section.add_student(student)

        # not all students were able to be assigned to a section, so we need to try to assign them to the sections with the least number of students and shuffle things around
        if len(unassigned_students) > 0:
            for student in unassigned_students:
                sections_sorted_by_size = sorted(course.sections, key=lambda x: len(x.students))
                
                assigned = False
                
                # try to assign the student to the section with the least number of students
                for min_size_section in sections_sorted_by_size:
                    if student.schedule[min_size_section.block] == None:
                        min_size_section.add_student(student)
                        assigned = True
                        break
                    if student.reassign_block(min_size_section.block, [min_size_section.block], 0):
                        if student.name == "Student 38 Grade 12" and give_output:
                            print("Reassigned waht they orriginaly had", student.name, " adding to section", min_size_section, "for course", course.name)
                        # print("Succesfuuly reassigned student", student.name, "to section", min_size_section.block)
                        min_size_section.add_student(student)
                        assigned = True
                        break
                # if the student is not able to be assigned to any of the sections, they cannot take this course
                if not assigned:
                    # print(f"Student {student.name} cannot be assigned to any sections for {course.name}.")
                    
                    # get student course preferences
                    for course_requests in student.course_requests:
                        has_alternative = False
                        try:
                            course_preference_rank = course_requests.index(course)
                            if course_preference_rank + 1 < len(course_requests):
                                has_alternative = True
                                next_course = course_requests[course_preference_rank + 1]
                            break
                        # haven't found the course in the list yet, so continue searching
                        except ValueError:
                            continue
                    
                    if not has_alternative:
                        continue
                        # print(f"Student {student.name} does not have any ranked alternative electives available.") 
                    else:
                        course.remove_student(student)
                        next_course.add_student(student)
                        courses_remaining.append(course)
        # if give_output and course.name == "MAT50":
        #     for section in course.sections:
        #         print("Section", section.block, "has", len(section.students), "students")
        #         for student in section.students:
        #             print(student.name, "is in block", student.schedule[section.block])
        if current_student is not None and current_student.schedule[section_to_assign.block] != section_to_assign:
            print(f"{current_student} NOT ASSIGNED TO THE section {section} THEY WERE SUPPOSED TO. Their schedule is {current_student.schedule}")
            raise Exception("Student not assigned to the correct section")
    while len(courses_remaining) > 0:
        courses_remaining = recursively_assign_students_using_max_flow(courses_remaining)
    
    return courses_remaining
def compare_students(s1: Student, s2: Student):
    s1_block_overlap = 0
    for course in s1.get_all_first_choice_courses():
        for section in course.sections:
            if section.block == section.block:
                s1_block_overlap += 1

    s2_block_overlap = 0
    for course in s2.get_all_first_choice_courses():
        for section in course.sections:
            if section.block == section.block:
                s2_block_overlap += 1
    
    if s1_block_overlap != s2_block_overlap:
        return s1_block_overlap - s2_block_overlap # if s1 has more overlap, it is less desirable to place in this block

    else:
        return s1.get_flexibility() - s2.get_flexibility() # students with less flexibility are more desirable to place in this block

def rollout(required_courses: list[Course], elective_courses: list[Course]):
    required_courses_sorted = sorted(required_courses, key=lambda x: len(x.sections))
    elective_courses_sorted = sorted(elective_courses, key=lambda x: len(x.sections))

    sorted_courses = required_courses_sorted + elective_courses_sorted
    
    all_courses_copy = deepcopy(sorted_courses)
    overall_best_cost = math.inf

    for i in range(len(all_courses_copy)):
        course = all_courses_copy[i]

        students_ranked = sorted(course.students, key=cmp_to_key(compare_students))
        for j in range(len(students_ranked)):
            # print("Student", j, "of", len(students_ranked), "for course", course.name, "iteration", i)
            student = students_ranked[j]

            best_state_section = None
            best_state_cost = math.inf

            current_test_state = deepcopy(all_courses_copy)
            best_test_state = None

            for k in range(len(course.sections)):
                # print("Testing section", k, "of", len(course.sections), "for student iteration", j, "for", course.name, "iteration", i)
                # rollout with copies of the original state
                section = current_test_state[i].sections[k]
                current_students_ranked = sorted(current_test_state[i].students, key=cmp_to_key(compare_students))
                current_student = current_students_ranked[j]
                
                
                if not section.is_full() and current_student.reassign_block(section.block, [section.block], 0):
                    # section.add_student(current_student)

                    # if k == 1 and student.name == "Student 42 Grade 9" and course.name == "MAT10":
                    #     print("OUTPUT FOR CASE WHERE COST IS 45 but the section is the same ast the best cost")
                    #     recursively_assign_students_using_max_flow(current_test_state[i:], give_output=True)
                    #     for course in current_test_state[:3]:
                    #         print(course.name, "has", len(course.students), "students")
                    #         for section in course.sections:
                    #             print("Section", section.block, "has", len(section.students), "students")
                    #             for student in section.students:
                    #                 print(student.name, "is in block", student.schedule[section.block])
                    #     MATH_10_COPY = deepcopy(current_test_state)
                    # else:
                    recursively_assign_students_using_max_flow(current_test_state, current_student, section)
                    cost = cost_of_final_schedules_from_courses(current_test_state)

                    # update the best state if the cost is lower
                    if cost < best_state_cost:
                        best_state_cost = cost
                        best_state_section = k
                        best_test_state = deepcopy(current_test_state)
                    # reset the courses to their original state
                    current_test_state = deepcopy(all_courses_copy)
                else:
                    print("Unable to assign. Is section full?", section.is_full(), "Is student already assigned?", current_student.get_course_section(course) != None, current_student.schedule[section.block])
            # assign the student based on the best result. TODO: check the case where the outcome is None
            if best_state_section == None:
                continue
                # print("Unable to assign student", j, "to any sections for", course.name)
            else:
                # what about the case where the student schedule was shuffled using reassign_block?
                best_section = course.sections[best_state_section]
                assigned = student.reassign_block(best_section.block, [best_section.block], 0)
                if not assigned:
                    print("Unable to assign student", student.name, "to section", best_section.block, "for course", course.name, "iteration", i)
                    continue
                if student.name == "Student 36 Grade 11":
                    print("Student", student.name, "is in block", best_section.block, "for course", course.name, "iteration", i)
                    print("Are they already place?", student.get_course_section(course))

                best_section.add_student(student)
                for i2 in range(len(student.schedule.values())):
                    for j2 in range(i2+1, len(student.schedule.values())):
                        if student.schedule[i2] != None and student.schedule[j2] != None and student.schedule[i2].name == student.schedule[j2].name:
                            print("Student", student.name, "is in the same course twice!", student.schedule)
                print("Best cost so far is ", best_state_cost)

                # test to see if this results in the same assignment as the original
                test_copy = deepcopy(all_courses_copy)

                # if compare_two_master_schedules(test_copy, best_test_state):
                #     print("The schedules are different. This is a problem.")
                #     raise ValueError("The schedules are different. This is a problem.")

            if best_state_cost <= overall_best_cost:
                overall_best_cost = best_state_cost
                overall_best_state = deepcopy(all_courses_copy)
                print("Overall best cost so far is ", overall_best_cost)
                test_overall_best = deepcopy(overall_best_state)
                print("Regaining best cost from the best state")
                recursively_assign_students_using_max_flow(test_overall_best, give_output=False)
                print("Best cost should be the same as before", cost_of_final_schedules_from_courses(test_overall_best))
                print("Cost of besttesttate", cost_of_final_schedules_from_courses(best_test_state))
                print("Comparing test_overall_best and best_test_state")
                print(compare_two_master_schedules(test_overall_best, best_test_state))
            elif best_state_cost > overall_best_cost:
                print("Best cost is worse than overall best cost. Stopping search.")
                print("Iteration number", i, "for course", course.name, "section", best_state_section, "cost", best_state_cost)
                print("Getting the best cost again from the best state")
                recursively_assign_students_using_max_flow(overall_best_state, give_output=True)
                best_cost = cost_of_final_schedules_from_courses(overall_best_state)
                print("Best cost is ", best_cost)
                # for course in overall_best_state[:3]:
                #     print(course.name, "has", len(course.students), "students")
                #     for section in course.sections:
                #         print("Section", section.block, "has", len(section.students), "students")
                #         for student in section.students:
                #             print(student.name, "is in block", student.schedule[section.block])

                print("******************8")
                print("Getting the best cost again from the current state")
                recursively_assign_students_using_max_flow(test_copy, give_output=True)
                # for course in test_copy[:3]:
                #     print(course.name, "has", len(course.students), "students")
                #     for section in course.sections:
                #         print("Section", section.block, "has", len(section.students), "students")
                #         for student in section.students:
                #             print(student.name, "is in block", student.schedule[section.block])
                # print("COMPARING MATH10 best and OVERALL BEST")
                # print(compare_two_master_schedules(overall_best_state, MATH_10_COPY))

                print("COMPARING AFTER ADDING EVERYTHING")
                print(compare_two_master_schedules(overall_best_state, test_copy))
                return
            
reset_student_assignments()
rollout(all_courses_required, all_courses_electives)

        
    
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

def visualize_schedule(people, n_people=10, person_type="Student"):
    fig, ax = plt.subplots(figsize=(10, 8))
    blocks = ['A', 'B', 'C', 'D', 'E', 'F', 'G']
    colors = ['#f4d03f', '#76d7c4', '#85c1e9', '#f1948a', '#bb8fce', '#FFA500', '#aed6f1']

    for block_idx, block in enumerate(blocks):
        # ax.add_patch(Rectangle((0, block_idx), 1, 1, color=colors[block_idx], label=f'{block} Block'))
        ax.text(-0.1, block_idx + 0.5, f'{block} Block', va='center', ha='right', fontsize=10, weight='bold')

    for person_idx, person in enumerate(people[:n_people]):  # Visualize only the first n students
        for block_idx, course in person.schedule.items():
            if course:
                ax.add_patch(Rectangle((person_idx, block_idx), 1, 1, color=colors[block_idx], alpha=0.7))
                ax.text(person_idx + .5, block_idx + 0.5, course.name, va='center', ha='center', fontsize=8)

    ax.set_xlim(0, len(people[:n_people]))
    ax.set_ylim(0, len(blocks))
    ax.set_xticks(range(1, len(people[:n_people]) + 1))
    ax.set_xticklabels([f'{person_type} {i+1}' for i in range(len(people[:n_people]))], rotation=45, ha='right')
    ax.set_yticks([])
    ax.set_aspect('equal')
    plt.tight_layout()
    plt.show()

# Call the function to visualize the schedule
```

```python
visualize_schedule(students)
```

```python
visualize_schedule(teachers, n_people=28, person_type="Teacher")
```
